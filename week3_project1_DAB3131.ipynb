```python
import numpy as np
import pandas as pd
Dataset = pd.read_csv('/content/Data-cleaning-for-beginners-using-pandas (1).csv')
avg = Dataset['Age'].median()
Dataset['Rating']=Dataset['Rating'].fillna(avg)
Dataset['Age']=Dataset['Age'].round().astype(object)
avg= Dataset['Rating'].median()
print(avg)
Dataset['Rating']=Dataset['Rating'].fillna(avg)
Dataset['Rating']=np.where(Dataset['Rating']==-1.0,"Unknown",Dataset['Rating'])
Dataset['Salary']=Dataset['Salary'].apply(lambda x:x.replace("$",""))
Dataset['Salary']=Dataset['Salary'].apply(lambda x:x.replace("k","000"))
year =Dataset[Dataset['Established']!= -1]
print(year)
Dataset['Established']=np.where(Dataset['Established']==-1,"Unknown",Dataset['Established'])
Dataset['Location']=Dataset['Location'].str.replace("Australia Aus","Australia,Aus")
Dataset['Location']=Dataset['Location'].str.replace("India In","India,In")
Dataset['Location']=Dataset['Location'].str.split("," ,expand=True).get(0)
Dataset['Easy Apply']=Dataset['Easy Apply'].apply(lambda y:y.replace("-1","FALSE"))

Dataset.head(29)
```
